```python
from scipy.spatial.distance import squareform, pdist
```

```python
data_df.loc[:, ['words_of_affirmation', 'quality_time',
       'receiving_gifts', 'acts_of_service', 'physical_touch', 'group']]
```

```python
distance_df = pd.DataFrame(squareform(pdist(data_df.loc[:, ['words_of_affirmation', 'quality_time',
       'receiving_gifts', 'acts_of_service', 'physical_touch', 'group']])), columns=data_df.name.unique(), index=data_df.name.unique())
```

```python
distance_df
```

### brute force assignment

```python
from random import shuffle
shuffled_names = list(distance_df.index)
```

```python
shuffle(shuffled_names)
shuffled_names
```

```python
assigned = set()
pairs = []
for subj in shuffled_names:
    ds = distance_df.to_dict()[subj]
    cur_max = 0
    candidate = subj
    if subj in assigned:
        continue
        
    for friend, score in ds.items():
        if friend in assigned:
            continue
            
        if score > cur_max:
            candidate = friend
    
    assigned.add(subj)
    assigned.add(candidate)
    pairs.append((subj, candidate))            
```
